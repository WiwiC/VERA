```python
import yt_dlp

# 1. Define the YouTube video URL and output file
url = "https://www.youtube.com/watch?v=gXwewPgLmkE"  # your video link
output_path = "TED_young_speaker.mp4"

# 2. Set yt_dlp options
ydl_opts = {
    'format': 'best[ext=mp4]+bestaudio[ext=m4a]/best',  # best video+audio, fallback to best
    'outtmpl': output_path,                              # output filename
    'merge_output_format': 'mp4'                         # ensure merged file is mp4
}

# 3. Download the video with error handling
try:
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])
    print(f"Video downloaded successfully as {output_path}")
except yt_dlp.utils.DownloadError as e:
    print("Download failed:", e)
```
